# Overview

We would like to build a model to predict whether one is going to be defaulted on loan.

## About the data set

- imbalanced

## Missing Values

    1. Dropping all the missing values.
        - concern: lossing many entries.
        - **However** Bagging from Random Forest may address this issue. 
    2. Median.
        - concern: may not accurate.
        - may not consider this option.

        - concern: regression assumptions are not met.

I am going to drop the missing values. 

The idea of bagging in Random Forest may address the concern of low training set. 

I will also perform other algorithms to compare their performance.

```python

df_dropped = df.dropna()
df_dropped.info()
```

## Undersampling

```python
df_default = df_dropped[df_dropped['BAD'] == 1]
df_nondefault = df_dropped[df_dropped['BAD'] == 0]

returns_underS = df_nondefault.sample(n=len(df_default), random_state=100)
df_underS = pd.concat([df_default, returns_underS])
df_underS
```

### Chi-Sqaure Test

#### 'REASON' 
It is binary variable. 

Getting dummy would not affect the total number of features. 

#### 'JOB'
It has 6 sub-categories. 

It would increase the number of features. 

See if I can just drop it by performing Chi-Sqaure

I would like to avoid the massive extension in number of columns due to the dummy variables.

- Null: the column has no relationship with the dependent variable. 
- Alt: the column has relationship with the dependent variable.

If it fails to reject the null, we say that the column has no relationship with the target variable (dependent varialbe)

Ideally, many categorical columns turn out to be having high p values so that I will fail to reject the Null (Drop the column)


```python
from scipy.stats import chi2_contingency

contTable = pd.crosstab(df_underS['JOB'],df['BAD'])
chi2_contingency(contTable)

```

```python
print("P value is {}".format(chi2_contingency(contTable)[1]))
print("Since P value is very small, we reject the null.")
print("It means that the column has relationship with the dependent variable")
```

```python
df_underS = pd.get_dummies(df_underS, columns = ['JOB','REASON'],drop_first = True)
df_underS
```

## Splitting Data

### into X and y

### into Training and Test


```python

X = df_underS.iloc[:,1:].values
y = df_underS.iloc[:,0].values

#Splitting the data into Training Set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)

#Normalizing the features
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

#Fitting Classifier to Training Set. Create a classifier object here and call it classifierObj
from sklearn.ensemble import RandomForestClassifier
classifierObj = RandomForestClassifier(criterion='entropy')
classifierObj.fit(X_train,y_train)

#Making predictions on the Test Set
y_pred = classifierObj.predict(X_test)

#Evaluating the predictions using a Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
print(classifierObj.score(X_test,y_test))
```

```python
TP = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TN = cm[1][1]

print("{} case(s) for 'we correctly predict that a client would default'".format(TP))
print("{} case(s) for 'we incorrectly predict that a client would default'".format(FP))
print("{} case(s) for 'we incorrectly predict that a client would not default'".format(FN))
print("{} case(s) for 'we correctly predict that a client would not default'".format(TN))
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision, recall, fscore, support = precision_recall_fscore_support(y_test, y_pred)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))
```
